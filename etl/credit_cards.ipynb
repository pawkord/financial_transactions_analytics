```python
df_cards = pd.read_csv('cards_data.csv')
df_cards.head()
```

```python
df_users = pd.read_csv('users_data.csv')
df_users.head()
```

```python
df = df.drop(columns=['errors'])
df['amount'] = df['amount'].replace('[\$,]', '', regex=True).astype(float)

df_cards['credit_limit'] = df_cards['credit_limit'].replace('[\$,]','', regex=True).astype(float)

cols_to_alter = ['per_capita_income', 'yearly_income', 'total_debt']
for col in cols_to_alter:
    df_users[col] = df_users[col].replace('[\$]', '', regex=True).astype(float)
```

```python
df.to_csv('transactions_c.csv', index=False)
df_cards.to_csv('cards_c.csv', index=False)
df_users.to_csv('users_c.csv', index=False)
```
